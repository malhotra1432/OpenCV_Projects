```python
# Reading video files
import cv2
import time


cap = cv2.VideoCapture('./video/my_video.mp4')

if cap.isOpened == False:
    print("Error! File Not Found! or wrong codec used")
    
while cap.isOpened:
    ret,frame = cap.read()
    if ret == True:
        # Writer 20 FPS 
        time.sleep(1/20)
        cv2.imshow('frame',frame)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break 
    else:       
        break

cap.release()
cv2.destroyAllWindows()
    
```
