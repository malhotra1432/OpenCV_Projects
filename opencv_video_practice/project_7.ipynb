# Corner Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('images/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess,cmap="gray")
```

```python
real_chessboard = cv2.imread("images/real_chessboard.jpg")
real_chessboard = cv2.cvtColor(real_chessboard,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chessboard)
```

```python
gray_real_chessboard = cv2.cvtColor(real_chessboard,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chessboard,cmap="gray")
```

```python
gray_flat_chess
```

```python
gray_real_chessboard
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst > 0.01 * dst.max()] = [255,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chessboard)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chessboard[dst > 0.01 * dst.max()] = [255,0,0]
```

```python
plt.imshow(real_chessboard)
```
